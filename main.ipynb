## Funciones

```python
def sigmoid(x):
    return 1 / ( 1 + np.exp(-x) )

# sigmoid = lambda x : 1 / ( 1 + np.exp(-x) )

def sigmoid_derivative(x):
    return x * (1 - x)

# sigmoid_derivative = lambda x : x * (1 - x)
```

## Dataset

```python
archivo = "dataset_frutas.csv"

# Leer datos del archivo
frutas : pd.DataFrame = pd.read_csv(archivo)
frutas
```

## One Hot Encoding

```python
# Color
frutas = pd.get_dummies(frutas, columns=["Color"])

frutas
```

```python
frutas.replace({True: 1, False: 0}, inplace=True)

frutas
```

## Reordenar columnas

```python
frutas.columns
```

```python
frutas = frutas.reindex(['Peso', 'Tamaño', 'Color_Amarillo',
       'Color_Claro', 'Color_Muy claro', 'Color_Muy oscuro', 'Color_Naranja',
       'Color_Oscuro', 'Color_Rojo', 'Color_Rojo claro', 'Color_Verde', 'Ácida', 'Dulce', 'Tropical'], axis=1)
frutas
```

```python
# Seleccionar las columnas a escalar (por ejemplo, 'columna1' y 'columna2')
columnas_a_escalar = ['Peso', 'Tamaño']

# Crear el scaler y aplicarlo solo a las columnas seleccionadas
scaler = StandardScaler()
frutas[columnas_a_escalar] = scaler.fit_transform(frutas[columnas_a_escalar])

frutas.head()
```

```python
training_inputs = np.array(frutas[['Peso', 'Tamaño', 'Color_Amarillo', 'Color_Claro', 'Color_Muy claro', 'Color_Muy oscuro', 'Color_Naranja', 'Color_Oscuro', 'Color_Rojo', 'Color_Rojo claro', 'Color_Verde']])
training_outputs = np.array(frutas[['Ácida', 'Dulce', 'Tropical']]).T

display(training_inputs)
display(training_outputs)
```

## Estructura

```python
input = training_inputs.shape[1]
input
```

```python
output = training_outputs.shape[0]
output
```

La estructura de la red son 11 neuronas de entrada y 3 de salida. No hay capas intermedias.

## Weights

Definiendo pesos random para partir de una base para hacer el entrenamiento. De tamaño tiene las variables input y output. Es una matriz de ancho (output) y alto (input).

```python
synaptic_weights = np.random.random((input, output))

synaptic_weights
```

## Funcion Fordward

Fordward propagation, avanza sobre la red primero calculando la multiplicacion punto entre el input (x) y el peso corrrespondiente al mismo (w). Por ultimo pasa por la funcion de activacion sigmoide.

```python
forward = lambda x, w: sigmoid(np.dot(x, w))
```

Ejemplo de fordward

```python
input_layer = training_inputs

# Genera un output haciendo la suma de los valores de input multiplicados por su peso sinaptico, y aplicandole la función sigmoide para normalizar a -1, 1
outputs = forward(input_layer, synaptic_weights)

outputs
```

```python
training_outputs = training_outputs.T
training_outputs
```

## Train

En una iteración de 25000 veces, genera un output aplicando la funcion fordward y la funcion sigmoide para la activacion de la misma. En base al error, se ajustan los pesos y se realizan las modificaciones a estos lo que sea necesario.

```python
epochs = 25000

for iteration in range(epochs):

    # Arranca con los valores de training
    input_layer = training_inputs

    # Genera un output haciendo la suma de los valores de input multiplicados por su peso sinaptico, y aplicandole la función sigmoide para normalizar a -1, 1
    outputs = forward(input_layer, synaptic_weights)

    # Calculo el error
    error = training_outputs - outputs

    # Ajusto los pesos en funcion al error y al resultado obtenido
    adjustments = error * sigmoid_derivative(outputs)
    synaptic_weights += np.dot(input_layer.T, adjustments)
```

## Backpropagation

En backpropagation, la idea es ajustar los pesos sinápticos en función del error calculado. Acá, se aplica la derivada de la función sigmoide a los valores de salida (sigmoid_derivative(outputs)) para medir la sensibilidad del error con respecto a los valores actuales de salida. Esto nos da una medida de cuánto ajustar los pesos en función del error y de la salida calculada. Esto se hace en la línea "adjustments = error * sigmoid_derivative(outputs)"

## Gradient Descent

Se usa gradient descent para actualizar los pesos (synaptic_weights) de la red. La actualización se hace sumando el producto de la matriz de las entradas rotadas (input_layer.T) y los ajustes calculados. Este es el paso que minimiza el error en cada iteración. Este paso se hace en la linea "synaptic_weights += np.dot(input_layer.T, adjustments)"
