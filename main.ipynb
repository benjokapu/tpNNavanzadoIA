```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import HTML, display
import tabulate
import networkx as nx
from random import random as rand
import itertools
```

## Funciones

```python
def sigmoid(x):
    return 1 / ( 1 + np.exp(-x) )

# sigmoid = lambda x : 1 / ( 1 + np.exp(-x) )

def sigmoid_derivative(x):
    return x * (1 - x)

# sigmoid_derivative = lambda x : x * (1 - x)
```

## Dataset

```python
archivo = "dataset_frutas.csv"

# Leer datos del archivo
frutas : pd.DataFrame = pd.read_csv(archivo)
frutas
```

## One Hot Encoding

```python
# Color
frutas = pd.get_dummies(frutas, columns=["Color"])

frutas
```

## Reordenar columnas

```python
frutas.columns
```

```python
frutas = frutas.reindex(['Peso', 'Tamaño', 'Color_Amarillo',
       'Color_Claro', 'Color_Muy claro', 'Color_Muy oscuro', 'Color_Naranja',
       'Color_Oscuro', 'Color_Rojo', 'Color_Rojo claro', 'Color_Verde', 'Ácida', 'Dulce', 'Tropical'], axis=1)
frutas
```

```python
training_inputs = np.array(frutas[['Peso', 'Tamaño', 'Color_Amarillo', 'Color_Claro', 'Color_Muy claro', 'Color_Muy oscuro', 'Color_Naranja', 'Color_Oscuro', 'Color_Rojo', 'Color_Rojo claro', 'Color_Verde']])
training_outputs = np.array(frutas[['Ácida', 'Dulce', 'Tropical']])

display(training_inputs)
display(training_outputs)
```

```python
training_values = np.concatenate((training_inputs.T, training_outputs.T)).T
table = lambda arr : display(HTML(tabulate.tabulate(arr, tablefmt='html')))
table(training_values)
```

```python
input = training_inputs.shape[1]
```

```python
output = training_outputs.shape[1]
```

```python
synaptic_weights = np.random.random((input, output))

synaptic_weights
```

```python
forward = lambda x, w: sigmoid(np.dot(x, w))
```

```python
input_layer = training_inputs

# Genera un output haciendo la suma de los valores de input multiplicados por su peso sinaptico, y aplicandole la función sigmoide para normalizar a -1, 1
outputs = forward(input_layer, synaptic_weights)

outputs
```
